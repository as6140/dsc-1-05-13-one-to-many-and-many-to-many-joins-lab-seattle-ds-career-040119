## Employees and their Office (a One-to-One join)

Return a list of all of the employees with their first name, last name and the city and state of the office that they work out of (if they have one). Include all employees and order them by their first name, then their last name.

```python
#Your code here
query = """
SELECT 
  employees.firstName
, employees.lastName
, offices.city
, offices.state
FROM employees 
JOIN offices
    ON employees.officeCode = offices.officeCode
ORDER BY
  employees.firstName DESC
, employees.lastName DESC
"""
#ORDER BY: ASC is default
#ORDER BY: DESC must be stated
```

```python
result = cur.execute(query)
```

```python
def py2_range(start,stop):
    """Return a list of all integers between start and stop."""
    result = []
    i = start
    while i < stop:
        result.append(i)
        i += 1
    return result
```

```python
py2_range(0,10)
```

```python
def py3_range(start,stop):
    """Yield each integer between start and stop."""
    i = start
    while i < stop:
        yield i
        i += 1
#return vs. yield:
#return says "we're done with this function, that's it"
#yield gives a generator object
```

```python
py3_range(0,10)
```

```python
range(0,10)
```

```python
for i in range(0,10):
    print(i)
```

## Customers and their Orders (a One-to-Many join)

Return a list of all of the customers. For each customer return a record for each of their order numbers, order dates and statuses.

```python
# Your code here
query = """
SELECT customers.customerNumber, orders.orderNumber, orders.orderDate, orders.status
FROM customers
JOIN orders
ON customers.customerNumber = orders.customerNumber
ORDER BY orders.status;
"""

pd.DataFrame(cur.execute(query).fetchall())
```
